```python
import pandas as pd
import random

# Eksempler for de to temaene
frukt_nouns = ["frukter", "epler", "pærer", "plommer", "bær"]
frukt_verbs_lg = ["plukker", "høster", "samler"]
frukt_verbs_pg = ["syltet", "kokte", "sorterte"]
reise_nouns = ["steder", "byer", "land", "områder"]
reise_verbs_lg = ["reiste", "dro", "flyttet"]
reise_verbs_pg_pret = ["bodde", "studerte", "arbeidet"]
reise_verbs_pg_inf = ["bo", "studere", "arbeide"]

# Lag eksempelsetninger
def generate_examples(n=5):
    examples = []
    for _ in range(n):
        # Frukt: pl-n-0 (parasitic før licensing), pret
        noun = random.choice(frukt_nouns)
        lg_verb = random.choice(frukt_verbs_lg)
        pg_verb = random.choice(frukt_verbs_pg)
        sent = f"Hvilke {noun} plukket alle som {pg_verb} dem {lg_verb}?"
        code = f"frukt pret pl-n-0"
        examples.append((sent, "frukt", "pret", "pl-n-0"))

        # Reise: pl-n-0 (parasitic før licensing), pret
        noun = random.choice(reise_nouns)
        lg_verb = random.choice(reise_verbs_lg)
        pg_verb = random.choice(reise_verbs_pg_pret)
        sent = f"Hvilke {noun} var det hun {lg_verb} til uten at hun {pg_verb} først?"
        code = f"reise pret pl-n-0"
        examples.append((sent, "reise", "pret", "pl-n-0"))

        # Reise: pl-n-0, inf
        noun = random.choice(reise_nouns)
        lg_verb = random.choice(reise_verbs_lg)
        pg_verb = random.choice(reise_verbs_pg_inf)
        sent = f"Hvilke {noun} var det hun {lg_verb} til uten å {pg_verb} først?"
        code = f"reise inf pl-n-0"
        examples.append((sent, "reise", "inf", "pl-n-0"))

    df = pd.DataFrame(examples, columns=["sentence", "keyword", "tense", "code"])
    return df

df_new_examples = generate_examples(5)

```

```python
import pandas as pd
import re

# Load the file
file_path = "/mnt/data/data-chr-2025 - Ark 1.csv"
df = pd.read_csv(file_path)

# Function to infer filler based on keyword
def infer_filler(keyword):
    keyword = keyword.lower()
    if keyword in ['epler', 'frukt', 'pærer', 'jordbær']:
        return 'dem'
    elif keyword in ['steder', 'byer', 'land']:
        return 'der'
    elif keyword in ['bøker', 'artikler']:
        return 'dem'
    else:
        return 'det'

# Apply the filler function
df['filler'] = df['nøkkelord'].apply(infer_filler)

# Update IDs from pl-n-0 to pl-n-1, pl-n-2, pl-n-3 etc.
def update_id(id_val, counter_dict):
    base = re.sub(r'\d+$', '', id_val)
    if base not in counter_dict:
        counter_dict[base] = 1
    else:
        counter_dict[base] += 1
    return f"{base}{counter_dict[base]}"

counter_dict = {}
df['id'] = df['id'].apply(lambda x: update_id(x, counter_dict))

import ace_tools as tools; tools.display_dataframe_to_user(name="Updated Parasitic Gap Data", dataframe=df)
```

```python
import pandas as pd
import random
from itertools import product
#import ace_tools as tools

# Eksempeldata for tematiske kategorier og strukturelle variasjoner
themes = {
    "frukt": {
        "objects": ["epler", "pærer", "plommer", "moreller", "jordbær"],
        "verbs_main": ["plukket", "sanket", "høstet", "samlet", "pakket"],
        "verbs_sub": ["kokte", "sylta", "rensa", "vaska", "kutta"],
    },
    "reise": {
        "objects": ["byer", "steder", "land", "hovedsteder", "reisemål"],
        "verbs_main": ["besøkte", "reiste til", "dro til", "var i", "utforska"],
        "verbs_sub": ["studerte", "undersøkte", "leste om", "skrev om", "hørte om"],
    }
}

structures = [
    ("pl-n-0", "før", "uten at hun {} først", "{}"),
    ("pl-n-1", "før", "uten å {} først", "{}"),
    ("lp-n-0", "etter", "fordi hun {} før hun {}", "{}"),
    ("lp-n-1", "etter", "fordi hun {} før hun {}", "{}"),
]

results = []

for theme, content in themes.items():
    for obj, verb_main, verb_sub, (code, tense, template, filler) in product(
        content["objects"], content["verbs_main"], content["verbs_sub"], structures):
        sentence = f"Hvilke {obj} {template.format(verb_sub, verb_main)}?"
        results.append({
            "key": theme,
            "tense": tense,
            "code": code,
            "sentence": sentence
        })

# Trekker ut 20 tilfeldige eksempler for variasjon
sampled_results = random.sample(results, 20)
df = pd.DataFrame(sampled_results)

```

```python
import pandas as pd
import random

# Eksempeldata med nøkkelord og utfyllende informasjon
keywords = [
    {'nøkkelord': 'frukt', 'filler': 'hvilke plommer', 'verb1': 'høstet', 'verb2': 'renset', 'adverb': 'før hun la dem i kurven', 'tempus': 'pret'},
    {'nøkkelord': 'reise', 'filler': 'hvilke byer', 'verb1': 'besøkte', 'verb2': 'studerte', 'adverb': 'før hun fikk stipend', 'tempus': 'pret'},
    {'nøkkelord': 'bøker', 'filler': 'hvilke bøker', 'verb1': 'kjøpte', 'verb2': 'leste', 'adverb': 'før hun skrev anmeldelsen', 'tempus': 'pret'},
    {'nøkkelord': 'filmer', 'filler': 'hvilke filmer', 'verb1': 'så', 'verb2': 'analyserte', 'adverb': 'før hun diskuterte dem i timen', 'tempus': 'pret'},
    {'nøkkelord': 'planter', 'filler': 'hvilke planter', 'verb1': 'plantet', 'verb2': 'vannet', 'adverb': 'før hun dro på ferie', 'tempus': 'pret'}
]

# Funksjon for å generere setninger
def generate_sentences(entry, order='lp', fill=0):
    f = entry['filler']
    v1 = entry['verb1']
    v2 = entry['verb2']
    adv = entry['adverb']
    t = entry['tempus']
    k = entry['nøkkelord']

    if order == 'lp':
        if fill == 0:
            s = f"{f} {v1} hun fordi hun {v2} {adv}?"
        elif fill == 1:
            s = f"{f} {v1} hun dem fordi hun {v2} {adv}?"
        elif fill == 2:
            s = f"{f} {v1} hun fordi hun {v2} dem {adv}?"
        elif fill == 3:
            s = f"{f} {v1} hun dem fordi hun {v2} dem {adv}?"
    elif order == 'pl':
        if fill == 0:
            s = f"{f} fordi hun {v2} {v1} hun {adv}?"
        elif fill == 1:
            s = f"{f} fordi hun {v2} dem {v1} hun {adv}?"
        elif fill == 2:
            s = f"{f} fordi hun {v2} {v1} hun dem {adv}?"
        elif fill == 3:
            s = f"{f} fordi hun {v2} dem {v1} hun dem {adv}?"
    code = f"{order}-n-{fill}"
    return {'kode': code, 'nøkkelord': k, 'setning': s, 'tempus': t}

# Generer setninger for alle kombinasjoner
data = []
for entry in keywords:
    for order in ['lp', 'pl']:
        for fill in [0, 1, 2, 3]:
            data.append(generate_sentences(entry, order, fill))

df = pd.DataFrame(data)
#import ace_tools as tools; tools.display_dataframe_to_user(name="Genererte setninger", dataframe=df)
```

```python
import pandas as pd
import random

# Define thematic verb-noun templates for English (lp = licensing before parasitic gap)
english_templates = [
    {
        "keyword": "fruit",
        "tense": "past",
        "base": "Which apples did she pick before she cleaned?",
        "gap2": "Which apples did she pick before she cleaned them?",
        "gap1": "Which apples did she pick them before she cleaned?"
    },
    {
        "keyword": "cities",
        "tense": "past",
        "base": "Which cities did she visit before she studied?",
        "gap2": "Which cities did she visit before she studied them?",
        "gap1": "Which cities did she visit them before she studied?"
    },
    {
        "keyword": "emails",
        "tense": "present",
        "base": "Which emails does he answer before he reads?",
        "gap2": "Which emails does he answer before he reads them?",
        "gap1": "Which emails does he answer them before he reads?"
    },
    {
        "keyword": "books",
        "tense": "present",
        "base": "Which books do they shelve before they categorize?",
        "gap2": "Which books do they shelve before they categorize them?",
        "gap1": "Which books do they shelve them before they categorize?"
    },
    {
        "keyword": "letters",
        "tense": "past",
        "base": "Which letters did he write before he mailed?",
        "gap2": "Which letters did he write before he mailed them?",
        "gap1": "Which letters did he write them before he mailed?"
    }
]

# Format into DataFrame
data = []
for i, tpl in enumerate(english_templates):
    base_code = f"lp-e-0"
    gap2_code = f"lp-e-2"
    gap1_code = f"lp-e-1"
    data.append([tpl["keyword"], tpl["tense"], base_code, tpl["base"]])
    data.append([tpl["keyword"], tpl["tense"], gap2_code, tpl["gap2"]])
    data.append([tpl["keyword"], tpl["tense"], gap1_code, tpl["gap1"]])

df_english = pd.DataFrame(data, columns=["keyword", "tense", "code", "sentence"])

#import ace_tools as tools; tools.display_dataframe_to_user(name="English PG Dataset", dataframe=df_english)
```
