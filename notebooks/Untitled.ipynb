```python
import pandas as pd

# Load the existing DataFrame
file_path = "parasittic-llama-openaigpt.xlsx"
df = pd.read_excel(file_path, index_col=0)

# Calculate absolute difference to expected_score for each model
df["diff_LLaMa"] = abs(df["LlaMa_score"] - df["expected_score"])
df["diff_gpt4"] = abs(df["gpt-4_score"] - df["expected_score"])
df["diff_gpt3.5"] = abs(df["gpt-3.5-turbo_score"] - df["expected_score"])


```
